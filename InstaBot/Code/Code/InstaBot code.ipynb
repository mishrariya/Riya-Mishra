```python
4. Follow/Unfollow given handle - 
   1. Open the Instagram Handle of “So Delhi”
   2. Start following it. Print a message if you are already following
   3. After following, unfollow the instagram handle. Print a message if you have already unfollowed.
```

```python
# Open Instagram profile of "So Delhi"
driver.get("https://www.instagram.com/sodelhi/")

# Wait for the profile page to load
time.sleep(2)

# Find the Follow button
follow_button = driver.find_element_by_xpath("//button[text()='Follow']")

# Check if already following
if follow_button.text == 'Follow':
    # Click on the Follow button to start following
    follow_button.click()
    print("You are now following So Delhi.")
else:
    print("You are already following So Delhi.")

# Wait for the action to complete
time.sleep(2)

# Find the Unfollow button
unfollow_button = driver.find_element_by_xpath("//button[text()='Following']")

# Check if already unfollowed
if unfollow_button.text == 'Following':
    # Click on the Unfollow button to unfollow
    unfollow_button.click()
    confirm_unfollow_button = driver.find_element_by_xpath("//button[text()='Unfollow']")
    confirm_unfollow_button.click()
    print("You have successfully unfollowed So Delhi.")
else:
    print("You have already unfollowed So Delhi.")
```

```python
5. Like/Unlike posts
   1. Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
   2. Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.
```

```python
# open Instagram profile of 'dilsefoodie'
driver.get("https://www.instagram.com/dilsefoodie/")

# Wait for the profile page to load
time.sleep(2)

# Find the first post
post = driver.find_element_by_xpath("(//div[contains(@class, 'v1Nh3')])[1]")
post.click()

# Wait for the post to load
time.sleep(2)

# Like the top 30 posts
for i in range(30):
    # Check if already liked
    like_button = driver.find_element_by_xpath("//span[@aria-label='Like']")
    if like_button.get_attribute("aria-label") == "Like":
        # Click on the Like button to like the post
        like_button.click()
        print("You have liked the post.")
    else:
        print("You have already liked the post.")

    # Move to the next post
    next_button = driver.find_element_by_xpath("//a[contains(@class, 'coreSpriteRightPaginationArrow')]")
    next_button.click()

    # Wait for the next post to load
    time.sleep(2)

# Unlike the top 30 posts
for i in range(30):
    # Check if already unliked
    unlike_button = driver.find_element_by_xpath("//span[@aria-label='Unlike']")
    if unlike_button.get_attribute("aria-label") == "Unlike":
        # Click on the Unlike button to unlike the post
        unlike_button.click()
        print("You have unliked the post.")
    else:
        print("You have already unliked the post.")

    # Move to the next post
    next_button = driver.find_element_by_xpath("//a[contains(@class, 'coreSpriteRightPaginationArrow')]")
    next_button.click()
```

```python
6. Extract list of followers
   1. Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.
   2. Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.
```

```python
driver.get("https://www.instagram.com/foodtalkindia/")

# Wait for the profile page to load
time.sleep(2)

# Find the followers button and click on it
followers_button = driver.find_element_by_xpath("//a[@href='/{}/followers/']".format('foodtalkindia'))
followers_button.click()

# Wait for the followers list to load
time.sleep(2)

# Scroll the followers list to ensure the desired number of followers is loaded
followers_list = driver.find_element_by_xpath("//div[@role='dialog']//ul")
followers_count = 0
while followers_count < 500:
    driver.execute_script("arguments[0].scrollIntoView();", followers_list)
    time.sleep(1)
    followers_count = len(followers_list.find_elements_by_tag_name("li"))
    
    # Break the loop if all the available followers are fetched
    if followers_count >= 500:
        break

# Extract the usernames of the first 500 followers
followers = []
for follower in followers_list.find_elements_by_tag_name("li")[:500]:
    username = follower.find_element_by_css_selector("a.FPmhX").text
    followers.append(username)

# Print the usernames of the first 500 followers of 'foodtalkindia'
print("Followers of 'foodtalkindia':")
for follower in followers:
    print(follower)

# Open Instagram profile of 'sodelhi'
driver.get("https://www.instagram.com/sodelhi/")

# Wait for the profile page to load
time.sleep(2)

# Find the followers button and click on it
followers_button = driver.find_element_by_xpath("//a[@href='/{}/followers/']".format('sodelhi'))
followers_button.click()

# Wait for the followers list to load
time.sleep(2)

# Scroll the followers list to ensure the desired number of followers is loaded
followers_list = driver.find_element_by_xpath("//div[@role='dialog']//ul")
followers_count = 0
while followers_count < 500:
    driver.execute_script("arguments[0].scrollIntoView();", followers_list)
    time.sleep(1)
    followers_count = len(followers_list.find_elements_by_tag_name("li"))
    
    # Break the loop if all the available followers are fetched
    if followers_count >= 500:
        break

# Extract the usernames of the first 500 followers
followers = []
for follower in followers_list.find_elements_by_tag_name("li")[:500]:
    username = follower.find_element_by_css_selector("a.FPmhX").text
    followers.append(username)

# Print the usernames of the first 500 followers of 'sodelhi'
print("\nFollowers of 'sodelhi':")
for follower in followers:
    print(follower)

# Open your own Instagram profile where sample username='abc'
driver.get("https://www.instagram.com/abc/")

# Wait for the profile page to load
time.sleep(2)

# Find the following button and click on it to view your following list
following_button = driver.find_element_by_xpath("//a[@href='/abc/following/']")
following_button.click()

# Wait for the following list to load
time.sleep(2)

# Scroll the following list to ensure all followers are loaded
following_list = driver.find_element_by_xpath("//div[@role='dialog']//ul")
following_count = len(following_list.find_elements_by_tag_name("li"))

# Extract the usernames of the followers you are following
following = []
for follower in following_list.find_elements_by_tag_name("li"):
    username = follower.find_element_by_css_selector("a.FPmhX").text
    following.append(username)


# Identify the followers of 'foodtalkindia' that you are following but who don't follow you back
non_followers = [username for username in following if username in followers]

# Print the followers of 'foodtalkindia' that you are following but who don't follow you back
print("Followers of 'foodtalkindia' that you are following but who don't follow you back:")
for username in non_followers:
    print(username)
```

```python
7. Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
   1. If You have already seen the story.
   2. Or The user has no story.
   3. Or View the story if not yet seen.
```

```python
# Open Instagram profile of 'coding.ninjas'
driver.get("https://www.instagram.com/coding.ninjas/")

# Wait for the profile page to load
time.sleep(2)

# Check if the story is available
story = driver.find_elements_by_xpath("//div[@role='dialog']//div[contains(@class, 'RR-M-')]")
if len(story) > 0:
    # Check if you have already seen the story
    seen_story = driver.find_elements_by_xpath("//div[@role='dialog']//div[contains(@class, 'VAVPi')]")
    if len(seen_story) > 0:
        print("You have already seen the story.")
    else:
        # View the story
        story[0].click()
        print("View the story.")
else:
    print("The user has no story.")
```
